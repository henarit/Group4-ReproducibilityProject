## Analyse results

```
import pickle 

# These are file paths for chart 3. 
with open('results/no_multi_mirna_clinical_gene.pkl', 'rb') as handle:
    results = pickle.load(handle)

with open('results/multi_mirna_clinical_gene.pkl', 'rb') as handle:
    results_multi = pickle.load(handle)
```

```
cancer_sites = ['BRCA', 'BLCA', 'CESC', 'COAD', 'READ', 'HNSC', 'KIRC', 'KIRP', \
                'LGG', 'LIHC', 'LUAD', 'LUSC', 'OV', 'PAAD', 'PRAD', 'SKCM', \
                'STAD', 'THCA', 'UCEC']

c = {} # c-index no multi
c_multi = {} # c-index multi
delta = {} # performance improvement

delta_cum = 0
for p in cancer_sites:
  # calculate average c-index for current cancer site p
  c[p] = sum(results[p])/len(results[p])
  c_multi[p] = sum(results_multi[p])/len(results_multi[p])

  # calculate average "relative performance improvement"
  # when using multimodal dropout
  if c[p] != 0:
    delta[p] = (c_multi[p]-c[p])/c[p]
  else: 
    delta[p] = 0
  delta_cum += delta[p]

# average improvement over all sites
delta['average'] = delta_cum/len(cancer_sites)

print('site', '\t\t', 'c-index', '\t', 'c-index multi', '\t', 'delta')
print('---------------------------------------------------------')
for p in sorted(cancer_sites):
  print(p, '\t\t', round(c[p], 2), '\t\t', round(c_multi[p], 2), '\t\t', \
        round(100*delta[p], 1), '%')
```
